```python
import requests
url='https://images.pexels.com/photos/36753/flower-purple-lical-blosso.jpg?auto=compress&cs=tinysrgb&dpr=2&h=650&w=940'
image_binary=requests.get(url).content
with open('flower.jpg','wb') as f:
    f.write(image_binary)
```

```python
from PIL import Image
flower_image=Image.open('flower.jpg')
```

```python
flower_image
```

```python
from pilkit.processors import Thumbnail

processor=Thumbnail(width=300,height=300)
```

```python
thumb_image=processor.process(flower_image)
```

```python
thumb_image
```

```python
thumb_image.save('flower_300_300.png')
thumb_image=thumb_image.convert('RGB')
for quality in [100,80,60,40,20]:
    thumb_image.save('flower_300_300-{}.jpg'.format(quality),quality=quality)
```
